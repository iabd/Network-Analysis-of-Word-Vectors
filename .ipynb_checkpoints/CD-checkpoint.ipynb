```python
def plotWithNodes(numberOfNodes, listCommunities):
    '''Plots the first community of the given number of nodes'''
    assert type(listCommunities) is list, "Please make sure that you pass the 'list' of communities!"
    assert type(numberOfNodes) is int, "please pass integer value of number of nodes"
    length=[len(y) for y in listCommunities]
    try:
        index=length.index(numberOfNodes)
    except:
        print("There's no community of given length")
    nx.draw(g.subgraph(listCommunities[index]), with_labels=True)
    plt.show()
```

```python
plotWithNodes(5, coms_gn)
```

```python
def getClique(g, k):
    kclique = list(community.k_clique_communities(g, k))
    kclique = [tuple(x) for x in kclique]
    print("Found "+ str(len(kclique)) + " cliques")
    return kclique
```

<h1>comparing partitions

```python
from sklearn.metrics import normalized_mutual_info_score
```

```python
def nmi_preprocess(partition):
    return [x[1] for x in sorted([(node,nid) for nid, cluster in enumerate(partition) for node in cluster], 
                                 key=lambda x: x[0])]
```

```python
lp_labels = nmi_preprocess(lp)
louvain_labels = nmi_preprocess(coms_louvain)
gn_labels = nmi_preprocess(coms_gn)
```

```python
normalized_mutual_info_score(lp_labels, louvain_labels)
```

```python
normalized_mutual_info_score(gn_labels, lp_labels)
```

```python
normalized_mutual_info_score(gn_labels, louvain_labels)
```

<h1> LFR: Synthetic networks with planted community structure

```python
from networkx.algorithms.community import LFR_benchmark_graph

n = 250 
tau1 = 3 
tau2 = 1.5 
mu = 0.1 


lgr_graph = LFR_benchmark_graph(n, tau1, tau2, mu, average_degree=5, min_community=20, seed=10)
```

```python
ground_truth_communities = [c for c in {tuple(m[1]['community']): None for m in lgr_graph.nodes.data()}]
```

```python
lp_coms_lfr = list(community.label_propagation_communities(lgr_graph))
lp_coms_lfr = [tuple(x) for x in lp_coms_lfr]
```

```python
lfr_gt = nmi_preprocess(ground_truth_communities)
lp_lfr = nmi_preprocess(lp_coms_lfr)
```

```python
normalized_mutual_info_score(lfr_gt, lp_lfr)
```

```python
nf = NF1(lp_coms_lfr, ground_truth_communities)
results = nf.summary()
```

```python
results['scores']
```

```python
results['details']
```

```python
nf.plot()
```
