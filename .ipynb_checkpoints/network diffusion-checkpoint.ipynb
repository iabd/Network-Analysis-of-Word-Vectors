```python
def plot_(plot, m, t):
    if plot==1:
        viz_ = DiffusionTrend(m, t)
        viz_.plot()
    elif plot==2:
        viz_ = DiffusionPrevalence(m, t)
        viz_.plot()
    else:
        print("wrong plot selected.. returning values")
```

```python
def thresholdModel(gr, plot=1,infected=0.3, threshold=0.1):
    '''Plots and returns the model and trend for threshold model. 
    Suppy plot=1 for DiffusionTrend and 2 for DiffusionPrevalence plots.
    '''
    model_ = th.ThresholdModel(gr)
    config_ = mc.Configuration()
    config_.add_model_parameter('percentage_infected', infected)
    for i in gr.nodes():
        config_.add_node_configuration("threshold", i, threshold)
    model_.set_initial_status(config_)
    trends_ = multi_runs(model_, execution_number=10, iteration_number=50, nprocesses=30)
    plot_(plot, model_, trends_)
    return model_, trends_
```

```python
modelth1, trendsth1=thresholdModel(g, plot=2)
```

```python
modelth1, trendsth1=thresholdModel(g, threshold=0.3, plot=2)
```

```python
modelther1, trendsther1=thresholdModel(ger, plot=2)
```

```python
modelther2, trendsther2=thresholdModel(ger, threshold=0.3, plot=2)
```

```python
modelthba1, trendsthba1=thresholdModel(gba, plot=2)
```

```python
modelthba2, trendsthba2=thresholdModel(gba, plot=2, threshold=0.3)
```

<h1> SIR

```python
def sirModel(gr, beta=0.05, gamma=0.05, infected=0.2, plot=1):
    '''Plots and returns the model and trend for SIR Model. 
    Suppy plot=1 for DiffusionTrend and 2 for DiffusionPrevalence plots.'''
    model_ = sir.SIRModel(gr)
    config_ = mc.Configuration()
    config_.add_model_parameter('beta', beta)
    config_.add_model_parameter('gamma', gamma)
    config_.add_model_parameter("percentage_infected", infected)
    model_.set_initial_status(config_)
    trends_ = multi_runs(model_, execution_number=10, iteration_number=100, nprocesses=30)
    plot_(plot, model_, trends_)
    return model_, trends_
```

<h1>SI

```python
def siModel(gr, beta=0.05, infected=0.2, plot=1):
    '''Plots and returns the model and trend for SI model. 
    Suppy plot=1 for DiffusionTrend and 2 for DiffusionPrevalence plots.'''
    model_ = si.SIModel(gr)
    cfg_ = mc.Configuration()
    cfg_.add_model_parameter('beta', beta)
    cfg_.add_model_parameter("percentage_infected", infected)
    model_.set_initial_status(cfg_)
    trends_ = multi_runs(model_, execution_number=10, iteration_number=100, nprocesses=30)
    plot_(plot, model_, trends_)        
    return model_, trends_
```

<h1>SIS

```python
def sisModel(gr, beta=0.05, lam=0.01, infected=0.2, plot=1):
    '''Plots and returns the model and trend for SIS model. 
    Suppy plot=1 for DiffusionTrend and 2 for DiffusionPrevalence plots.'''
    model_ = sis.SISModel(gr)
    cfg_= mc.Configuration()
    cfg_.add_model_parameter('beta', beta)
    cfg_.add_model_parameter("percentage_infected", infected)
    cfg_.add_model_parameter('lambda', lam)
    model_.set_initial_status(cfg_)
    trends_ = multi_runs(model_, execution_number=10, iteration_number=100, nprocesses=30)
    plot_(plot, model_, trends_)        
    return model_, trends_
```
